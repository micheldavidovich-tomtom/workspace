# Preprocessing data

Building a set of functions that joins the data of the counts of each point to the API responses data.

```
def deduplicate_on(df, on_column, counts_column) -> pd.DataFrame:
    """
    """
    df_counts = df.groupby([on_column])[counts_column].sum().reset_index()
    df_to_compare = df.drop_duplicates(subset=[on_column], keep='first').drop(counts_column, axis=1)
    return df_to_compare.merge(df_counts, how='right', on=on_column)

def preprocess_addresses(original_path, api_responses_path) -> pd.DataFrame:
    """
    """
    original_df = pd.read_parquet(original_path)
    original_df = deduplicate_on(original_df, 'searched_query', 'search_query_counts')
    
    api_resp_df = pd.read_parquet(api_responses_path).drop_duplicates(subset=['searched_query'], keep='first')

    result = api_resp_df.merge(
        original_df[['searched_query', 'search_query_counts']], how='left', left_on=['searched_query'], right_on=['searched_query']
    )
    
    return result[result['searched_query'].str.lower() != 'structuredGeocode.JSON'.lower()]

def enrich_and_deduplicate_responses(df_comp, df_incomp) -> pd.DataFrame:
    """
    """
    df_full = pd.concat([df_comp, df_incomp])
    return deduplicate_on(df_full, 'searched_query', 'search_query_counts')
  
def impose_geobias(df, polygon_path, sep: str = ';', header: bool = True, crs: str = "EPSG:4326") -> pd.DataFrame:
    """
    """
    df_copy = df.copy()
    df_copy['coordinates'] = df_copy.apply(lambda x: Point(x['lon'], x['lat']), axis=1)
    df_coords = gpd.GeoDataFrame(df_copy, geometry='coordinates', crs=crs)
    
    polygon_df = pd.read_csv(polygon_path, sep=sep)
    polygon_shape = gpd.GeoSeries.from_wkt(polygon_df['geom'])
    geo_df = gpd.GeoDataFrame(polygon_df, geometry=polygon_shape, crs=crs)
    
    return gpd.sjoin(df_coords, geo_df, how='inner')

def preprocess_all_responses(
    original_path_comp, api_responses_path_comp, original_path_incomp, api_responses_path_incomp, polygon_path, sep: str = ';'
) -> pd.DataFrame:
    """
    """
    df_comp = preprocess_addresses(original_path_comp, api_responses_path_comp)
    df_incomp = preprocess_addresses(original_path_incomp, api_responses_path_incomp)
    df_preprocessed = enrich_and_deduplicate_responses(df_comp, df_incomp)
    
    return impose_geobias(df_preprocessed, polygon_path, sep)
```

## Exploding responses

```
def explode_points(df: pd.DataFrame, capped_counts: int = 100, count_column: str = 'search_query_counts') -> pd.DataFrame:
    """Function that multiples the points in a DataFrame by the number of occurences of the point based on a count column.

    :param df: DataFrame that has the counts for searched queries
    :type df: pd.DataFrame
    :param capped_counts: Max number of counts a searched query will have. Defaults to 100
    :type capped_counts: int
    :param count_column: The column where the DataFrame contains the counts for each searched query.
    :type count_column: str
    :return: A DataFrame that has one row for each time a searched query appears, instead of one row with the searched query and the count. Defaults to 'search_query_counts'
    :rtype: pd.DataFrame
    """
    df_copy = df.copy()
    # Keeping only those addresses with less than 2x the selected max repetitions
    df_copy = df_copy[df_copy[count_column] < 2 * capped_counts]
    # Capping the values that are over the repetitions to capped_counts
    df_copy.loc[df_copy[count_column] > capped_counts, count_column] = capped_counts
    df_final = pd.DataFrame()
    for _, row in df_copy.iterrows():
        real_num_rows = pd.concat([pd.DataFrame([row.values], columns=row.index)] * row[count_column])
        df_final = pd.concat([df_final, real_num_rows])

    return df_final
```
